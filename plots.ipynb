```python
import numpy as np
import matplotlib.pyplot as plt
from read_write import quadrature_and_phase
```

```python
Ex_mod, Ex_phase = quadrature_and_phase("./dados/realEx.datab", "./dados/imagEx.datab")
```

```python
x = np.arange(-10000, 10250, 250)
```

```python
plt.style.use('default')
```

```python
arra = np.array(Ex_mod, dtype=np.float64)
freq = []
for i in range(0, 4):
    freq += [np.log(arra[i*81:(i+1)*81], dtype=np.float64)]
```

```python
def cm_to_inche(value):
    return value/2.54
```

```python
plt.figure(figsize=(cm_to_inche(26),cm_to_inche(13)))
plt.plot(x, freq[0], 'o', color='red', label='0.25 Hz')
plt.plot(x, freq[1], 'o', color='blue', label='0.75 Hz')
plt.plot(x, freq[2], 'o', color='orange', label='1.25 Hz')
plt.plot(x, freq[3], 'o', color='green', label='3.00 Hz')
plt.legend()
plt.grid(True)
plt.xlabel("$x$ (m)")
plt.ylabel("$\log |E_x|$ (V/m)")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(cm_to_inche(26), cm_to_inche(13)))

axs[0, 0].plot(x, freq[0], '.', color='red', label='0.25 Hz')
axs[0, 0].legend()
axs[0, 0].grid(True)
axs[0, 0].set_xlabel("$x$ (m)")
axs[0, 0].set_ylabel("$\log |E_x|$ (V/m)")

axs[1, 0].plot(x, freq[1], '.', color='blue', label='0.75 Hz')
axs[1, 0].legend()
axs[1, 0].grid(True)
axs[1, 0].set_xlabel("$x$ (m)")
axs[1, 0].set_ylabel("$\log |E_x|$ (V/m)")

axs[0, 1].plot(x, freq[2], '.', color='orange', label='1.25 Hz')
axs[0, 1].legend()
axs[0, 1].grid(True)
axs[0, 1].set_xlabel("$x$ (m)")
axs[0, 1].set_ylabel("$\log |E_x|$ (V/m)")

axs[1, 1].plot(x, freq[3], '.', color='green', label='3.00 Hz')
axs[1, 1].legend()
axs[1, 1].grid(True)
axs[1, 1].set_xlabel("$x$ (m)")
axs[1, 1].set_ylabel("$\log |E_x|$ (V/m)")
fig.tight_layout()
```
